### Some Summary Stats

```python
nRows = training.shape[0]
nDups = sum(training["is_duplicate"])
nWords1 = sum(training["question1"].apply(lambda question: len(str(question).split(" "))))
nWords2 = sum(training["question2"].apply(lambda question: len(str(question).split(" "))))

print("Total number of observations: {0}".format(nRows))
print("Proportion of duplicate posts: {0:.4f}".format(nDups/nRows))
print("Mean words in question 1: {0:.4f}".format(nWords1/nRows))
print("Mean words in question 2: {0:.4f}".format(nWords2/nRows))
```
